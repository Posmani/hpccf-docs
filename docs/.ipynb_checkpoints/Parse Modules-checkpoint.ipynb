```python
from github import Github
import os
```

```python
g = Github(os.getenv('GH_PAT'))
```

```python
repo = g.get_repo('ucdavis/spack-ucdavis')
```

```python
contents = repo.get_contents('modulefiles/hpccf/franklin')
modulefiles = []
while contents:
    file_content = contents.pop(0)
    if file_content.type == "dir":
        contents.extend(repo.get_contents(file_content.path))
    else:
        if file_content.name.endswith('.lua'):
            modulefiles.append(file_content)
            print(f'Found {file_content.path}')
```
